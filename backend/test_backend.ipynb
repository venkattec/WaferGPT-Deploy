```python
import os
import shutil
import random

def copy_images_from_subfolders(root_folder: str, target_folder: str):
    # Check if target folder exists, create it if it doesn't
    if not os.path.exists(target_folder):
        os.makedirs(target_folder)
    
    # Initialize a counter for naming images
    image_counter = 1
    
    # Iterate over each subfolder inside the root folder
    for subfolder in os.listdir(root_folder):
        subfolder_path = os.path.join(root_folder, subfolder)
        
        # Skip if it's not a directory
        if not os.path.isdir(subfolder_path):
            continue
        
        # List all images in the subfolder (assuming image files have common image extensions)
        images = [f for f in os.listdir(subfolder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff'))]
        
        if images:
            # Randomly pick one image from the subfolder
            selected_image = random.choice(images)
            selected_image_path = os.path.join(subfolder_path, selected_image)
            
            # Define the target path with the naming format sem1.png, sem2.png, ...
            target_image_name = f"sem{image_counter}.png"
            target_image_path = os.path.join(target_folder, target_image_name)
            
            # Copy the selected image to the target folder with the new name
            shutil.copy(selected_image_path, target_image_path)
            
            # Increment the image counter for the next image
            image_counter += 1

    print(f"Images have been copied to '{target_folder}' with sequential names.")

# Example Usage:
root_folder = "/home/sbna/Documents/SEM/master_dataset_non_defect/train/"  # Provide your folder path here
target_folder = "/home/sbna/Documents/WaferGPT-Frontend/sem_images"  # This folder will be created to store the images

copy_images_from_subfolders(root_folder, target_folder)
```
